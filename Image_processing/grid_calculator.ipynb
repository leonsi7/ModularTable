```python
from skimage import io, color, filters, feature, measure
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np

# Charger l'image et l'afficher
image = io.imread("data/1.jpg")
plt.imshow(image, cmap='gray')
```

```python
# Convertir l'image en niveaux de gris
gray_image = color.rgb2gray(image)
plt.imshow(gray_image, cmap='gray')
```

```python
# Appliquer le filtre Canny pour détecter les contours
edges = feature.canny(gray_image)
plt.imshow(edges,cmap='gray')
```

```python
projection_x = np.sum(edges, axis=0)

projection_y = np.sum(edges, axis=1)
plt.plot(projection_x)
```
